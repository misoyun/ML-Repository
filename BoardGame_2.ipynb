```python
import pandas as pd

raw_data = pd.read_csv("../dataSet/board-game-data/bgg_db_1806.csv", index_col = 'rank')
display(raw_data)
```

```python
import seaborn as sns
display(sns.distplot(raw_data['avg_rating']))
```

```python
display(raw_data.isnull().sum())
```

```python
display(raw_data['avg_rating'].describe())
```

```python
raw_data.drop(['bgg_url', 'game_id', 'names', 'image_url', 'num_votes', 'owned'], axis = 1, inplace = True)
```

```python
display(raw_data)
```

```python
data_cleaner = [raw_data]
for data in data_cleaner:
    data['success'] = pd.cut(data['avg_rating'], 2,labels = False,duplicates = 'drop')

display(sns.distplot(raw_data['success']))
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(raw_data, raw_data.success, test_size = 0.3, random_state = 1)
display(X_test.info())
```

```python
data_cleaner = [X_train, X_test]
for data in data_cleaner:
    data['avg_players'] = (data['min_players'] + data['max_players'])/2.
    data['mechanic_1'] = data['mechanic'].str.split(", ", expand = True)[0]
    data['mechanic_2'] = data['mechanic'].str.split(", ", expand = True)[1]
    data['mechanic_3'] = data['mechanic'].str.split(", ", expand = True)[3]
    data['mechanic_4'] = data['mechanic'].str.split(", ", expand = True)[4] 
    
    data['category_1'] = data['category'].str.split(", ", expand = True)[0]
    data['category_2'] = data['category'].str.split(", ", expand = True)[1]
    
    data['designer_1'] = data['designer'].str.split(", ", expand = True)[0]

display(X_train)
```
